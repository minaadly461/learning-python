```python
weight = float(input("Enter your weight: "))
unit = input("Killograms or pounds? (K or L )")

if unit =='K':
  weight = weight * 2.205
  unit ='Lbs.'
  print(f'your weight is : {weight}{unit}')

elif unit =='L':
  weight = weight / 2.205
  unit = 'Kgs.'
  print(f'your weight is : {weight}{unit}')

else : 
  print(f'{unit} you have entered is not valid')
```
