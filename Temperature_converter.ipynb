```python
temp = float(input("Enter the temperatur"))
unit = input("is this temperature in celsius or fehrenheite (C or F):")

if unit == "C":
  temp = (9 * temp) / 5 + 32
  print(f'the temperature in fehrenheite is : { temp} F')
elif unit == "F":
  temp = (temp - 32) * 5 / 9
  print(f'the temperature in Celsius is : {temp} C')
else :
  print(f'this {unit} is not valid please choose from (C or F)')
```
