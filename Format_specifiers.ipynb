# Format specifiers

```python
number_1 = 1234.56789
number_2 = -9880.23
number_3 = 34.5
```

```python
print(f"number 1 :{number_1:.2f}")
print(f"number 1 :{number_2:.2f}")
print(f"number 1 :{number_3:.2f}")
```

```python
print(f"number 1 :{number_1:10}")
```

```python
print(f"number 1 :{number_1:+}")
```

```python
print(f"number 2 :{number_2:=}")
```

```python
print(f"number 1 :{number_1:,}")
```
