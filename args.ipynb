```python
# to solve this we use arpitary argument


def add(*args):
  total = 0
  for arg in args :
    total +=arg

  return total


print(add(1,2,3))
```

```python
def display_name(*args):
  for arg in args :
    print(arg , end=" ")


display_name("mina","adly","lahzy")
```

```python
def print_address(**kwargs): # * return tuple but ** return dectionary
  for k , v in kwargs.items():
    print(k , v)

print_address(apr_num = 1 , street="alnasr" , city="cairo" , country="Egypt")
```

```python
def shipping_label(*args , **kwargs):
  for arg in args:
    print(arg , end=' ')

  print()

  for k , v in kwargs.items():
    print(k , v , end = " - ")

shipping_label("Dr","mina","adly","lahzy",
               apr_num = 1 , street="alnasr" , city="cairo" , country="Egypt")
```
