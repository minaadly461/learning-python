```python
def net_price(item_price , discount=0, tax = .14):
  return item_price * (1-discount) * (1 + tax)


print(f"price is {net_price(500 ):.2f} l.e")
```

```python
import time 


def count(end , start = 0):
  for i in range(start , end + 1):
    print(i)
    time.sleep(1)
  print("Done !")


count(8)
```
