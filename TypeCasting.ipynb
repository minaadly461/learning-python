# Type Casting

```python
full_name = "mina Adly "
age = 21 
gpa = 3.5 
is_sutdent = True
```

```python
print(type(is_sutdent))
```

```python
gpa = int(gpa)
print(gpa)
```

```python
age = float(age)
print(age)
```

```python
full_name = bool(full_name)

print(full_name)
```

```python
is_sutdent = int(is_sutdent)
print(is_sutdent)
```

```python
age = str(age)
type(age)
```

```python
print(age)
```
